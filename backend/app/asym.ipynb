```python
def distance_asymmetric_binary(i,j):
    #measure for asymmetric binary is d(i,j) = b+c / a+b+c
    a = 0 
    b = 0
    c = 0
    for num in range(len(i)):
        index = num
        if (i[index] == 1 and j[index] == 0):
            b = b+1
        elif (i[index] == 1 and j[index] == 1):
            a = a+1
        elif (i[index] == 0 and j[index] == 1):
            c = c+1
    up = b+c
    down = a+b+c
    distance = up/down
    return(i,j,a,b,c,distance)
```

```python
recommend_flower = matrix_result_flower
recommend_leaf = matrix_result_leaf
recommend_cactus = matrix_result_cactus
print(recommend_cactus)
```

```python
def sort_ranking(matrix_distance):
    matrix_rank1 = []
    matrix_rank2 = []
    matrix_rank3 = []
    #matrix_rank4 = []
    #matrix_rank5 = []
    for rank in range(5):
        q = 'index == '+str(rank)
        rank_dis = matrix_distance.query(q).values.tolist()
        tree = rank_dis[0][0]
        if (rank == 0):
            matrix_rank1.append(tree)
        elif (rank == 1):
            matrix_rank2.append(tree)
        elif (rank == 2):
            matrix_rank3.append(tree)
        #elif (rank == 3):
            #matrix_rank4.append(tree)
        #elif (rank == 4):
            ##matrix_rank5.append(tree)
    return(matrix_rank1,matrix_rank2,matrix_rank3)
```

```python
def DataAndSample(type, data_matrix,testing_sampling):
    dist = []
    run_cactus = 0
    check = 0

    print('data testing : '+str(testing_sampling))
    
    for dat in data_matrix.index:
        que = 'index == '+str(dat)
        data = data_matrix.query(que).values.tolist()
        datasum = sum(data,[])
        print(str(datasum))
        predict = distance_asymmetric_binary(testing_sampling,datasum)
        d = [predict[5]]
        dist.append(d)

    dist = pd.DataFrame(data=dist, columns=['distance'])
    if (type == 0):
        #flower type
        matrix_distance = pd.concat([recommend_flower, dist], axis=1)
    elif (type == 1):
        #leaf type
        matrix_distance = pd.concat([recommend_leaf, dist], axis=1)
    else :
        #cactus type
        matrix_distance = pd.concat([recommend_cactus, dist], axis=1)

    matrix_distance = matrix_distance.sort_values(ascending=True,by=['distance'],ignore_index=True)
    ranking = sort_ranking(matrix_distance)
    print(matrix_distance)  
    print(ranking)
    return(ranking)
```

```python
def hitRatio(matrix):
    matrix['rank1'] = matrix['rank1'].astype('int')
    matrix['rank2'] = matrix['rank2'].astype('int')
    matrix['rank3'] = matrix['rank3'].astype('int')
    if (matrix['index_tree'].values in matrix['rank1'].values ):
        matrix.loc[matrix['index_tree'].values == matrix['rank1'].values , 'hit rate'] = 1
    if (matrix['index_tree'].values in matrix['rank2'].values ):
        matrix.loc[matrix['index_tree'].values == matrix['rank2'].values , 'hit rate'] = 1
    if (matrix['index_tree'].values in matrix['rank3'].values ):
        matrix.loc[matrix['index_tree'].values == matrix['rank3'].values , 'hit rate'] = 1
    else :
        matrix.loc[matrix['hit rate'] == 'NaN' ] = 0
    #matrix.loc[matrix['index_tree'].values == matrix['rank2'].values , 'hit rate'] = 1
    #matrix.loc[matrix['index_tree'].values == matrix['rank3'].values , 'hit rate'] = 1
    #matrix.loc[matrix['index_tree'].values != matrix['rank1'].values , 'hit rate'] = 0
    #matrix.loc[matrix['index_tree'].values != matrix['rank2'].values , 'hit rate'] = 0
    #matrix.loc[matrix['index_tree'].values != matrix['rank3'].values , 'hit rate'] = 0
    hitrate = matrix['hit rate'].value_counts()
    #hitrate = hitrate.loc['1.0']
    #print(matrix)
    #print(hitrate) 
    return matrix,hitrate
```
